```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x= tips.sex, y=tips.total_bill, data= tips)
```

```python
sns.barplot(x= tips.sex, y=tips.total_bill, data= tips, estimator="sum")
```

```python
sns.barplot(x= tips.sex, y=tips.total_bill, data= tips, estimator= np.std)
```

```python
sns.countplot(x=tips.sex, data=tips)
```

```python
sns.boxplot(x='day', y=tips.total_bill, data=tips)
```

```python
sns.boxplot(x='day', y=tips.total_bill, data=tips, hue=tips.smoker)
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.smoker)
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, split=True)
```

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, jitter=True)
```

```python
sns.stripplot(x=tips.day, y=tips.total_bill, data=tips, jitter=True, hue=tips.sex)
```

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex)
```

```python
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips)
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
sns.catplot(x=tips.day, y=tips.total_bill, data=tips, kind= 'bar')
```

```python
sns.catplot(x=tips.day, y=tips.total_bill, data=tips, kind= 'violin')
```
