```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
        'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen',
        'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
        'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol',
        'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

### Linear regressioin model###



```python
targets = data_preprocessed.log_price

inputs = data_preprocessed.drop(['log_price'], axis= 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
input_scaled = scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_scaled, targets, test_size=0.2, random_state=365)
```

```python
reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
```

```python
sns.displot(y_train-y_hat)
plt.title('Resduals PDF', size=18)
```

```python
reg.score(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['weights'] = reg.coef_
reg_summary
```

```python
df_pf = pd.DataFrame(y_hat_test, columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)

df_pf['Target'] = df_pf['Target'].round(2)

df_pf.head(15)
```
