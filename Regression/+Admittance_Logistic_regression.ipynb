```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('/content/2.01.+Admittance (1).csv')
```

```python
raw_data.head()
```

```python
data = raw_data.copy()
data['Admitted'] = data['Admitted'].map({'Yes':1, 'No':0})
data.head()
```

```python
y = data.Admitted
x1 = data.SAT
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y, x)
result_log = reg_log.fit()
```

```python
result_log.summary()
```
