```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/1.01.+Simple+linear+regression (2).csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
x = data['SAT']
y = data['GPA']
```

```python
x.shape
```

```python
x_matrix = x.values.reshape(84, 1)
```

```python
y_matrix = y.values.reshape(84, 1)
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_matrix, y_matrix)
```

```python
reg.score(x_matrix, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
input_data = np.array([1740])
input_data_reshaped = input_data.reshape(1, -1)
```

```python
reg.predict(input_data_reshaped)
```

```python
reg.predict([[1740]])
```

```python
new_data = pd.DataFrame(data=[1740, 1760], columns=['SAT'])
new_data
```

```python
reg.predict(new_data)
```

```python
new_data['Predicted_GPA'] = reg.predict(new_data)
```

```python
new_data
```

```python
plt.scatter(x,y)

yhat = reg.coef_ * x_matrix + reg.intercept_

fig = plt.plot(x, yhat, lw=2, c='#006837')

plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
```
