```python
import yfinance as yf
```

```python
start = datetime(2006, 1, 1)
```

```python
end = datetime(2016, 1, 1)
```

```python
# tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
# start_date = '2006-01-01'
# end_date = '2024-03-29'

# bank_stock_data = yf.download(tickers, start=start_date, end=end_date)

# bank_stock_data.head()
```

```python
BAC = yf.download('BAC', start=start, end=end)

C = yf.download('C', start=start, end=end)

GS = yf.download('GS', start=start, end=end)

JPM = yf.download('JPM', start=start, end=end)

MS = yf.download('MS', start=start, end=end)

WFC = yf.download('WFC', start=start, end=end)
```

```python
BAC.head()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stock = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
bank_stock.head()
```

```python
bank_stock.drop(['Adj Close'], level=1, axis=1, inplace=True)
```

```python
bank_stock.columns.names = ['Bank Tiker', 'Stock Info']
```

```python
bank_stock.head()
```

```python
for tick in tickers:
  print(tick,'--',bank_stock[tick]['Close'].max())
```

```python
bank_stock.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
  returns[tick + ' '+'Return'] = bank_stock[tick]['Close'].pct_change()
```
