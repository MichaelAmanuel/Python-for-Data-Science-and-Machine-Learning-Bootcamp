```python
data = dict(
    type='choropleth',
    locations=['AZ', 'CA', 'NY'],
    locationmode='USA-states',
    colorscale='Portland',
    text=['Arizona', 'California', 'New York'],
    z=[1.0, 2.0, 3.0],
    colorbar={'title': 'Colorbar Title Goes Here'})
```

```python
layout = dict( geo={'scope':'usa'})
```

```python
choromap = go.Figure(data=[data], layout=layout)
```

```python
df = pd.read_csv('/content/2011_US_AGRI_Exports')
```

```python
df.head()
```

```python
import plotly.graph_objects as go

data = dict(
    type='choropleth',
    colorscale='YlOrRd',
    locations=df['code'],
    locationmode='USA-states',
    z=df['total exports'],
    text=df['text'],
    marker=dict(line=dict(color='rgb(255,255,255)', width=2)),
    colorbar=dict(title='Millions USD')
)

layout = dict(title='2011 US Agriculture Exports by State',
              geo=dict(scope='usa', showlakes=True, lakecolor='rgb(85,173,240)'))

choromap2 = go.Figure(data=[data], layout=layout)
```

```python
iplot(choromap2)
```

```python
df1 = pd.read_csv('/content/2014_World_GDP')
```

```python
df1.head()
```

```python
data = dict(
            type='choropleth',
            locations=df1['CODE'],
            z=df1['GDP (BILLIONS)'],
            text=df1['COUNTRY'],
            colorbar=dict(title='GDP (BILLIONS)')
            )

layout = dict( title='2014 Global GDP',
              geo=dict(showframe=False, projection_type='mercator')
              )

choromap3 = go.Figure(data=[data], layout=layout)
```
