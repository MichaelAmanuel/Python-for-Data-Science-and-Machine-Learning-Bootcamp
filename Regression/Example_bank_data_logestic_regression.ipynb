```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('/content/Example_bank_data (1).csv')
```

```python
raw_data.head()
```

```python
raw_data.set_index('Unnamed: 0', inplace=True)
```

```python
raw_data.head()
```

```python
data = raw_data.copy()
data['y'] = data['y'].map({'yes':1, 'no':0})
data.head()
```

```python
y = data.y
x1 = data.duration
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y, x)
result_log = reg_log.fit()
```

```python
result_log.summary()
```
